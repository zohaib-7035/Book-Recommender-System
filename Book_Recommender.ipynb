```python
data=rating.merge(book,on='ISBN')
```

```python
num_rating=data.groupby('title')['Rating'].count().reset_index()
```

```python
num_rating.rename(columns={'Rating':'num_of_rating'},inplace=True)
```

```python
final=data.merge(num_rating,on='title')
```

```python
final=final[final['num_of_rating']>=50]
```

```python
final.head()
```

```python
final.drop_duplicates(['ID','title'],inplace=True)
```

```python
book_pivot=final.pivot_table(columns='ID',index='title',values='Rating')
```

```python
book_pivot
```

```python
book_pivot.shape
```

```python
book_pivot.fillna(0,inplace=True)
```

```python
book_pivot
```

```python
from scipy.sparse import csr_matrix
book_sparse=csr_matrix(book_pivot)
```

```python
book_sparse
```

```python
from sklearn.neighbors import NearestNeighbors
model=NearestNeighbors(algorithm='brute')
```

```python
model.fit(book_sparse)
```

```python
distance, suggestion=model.kneighbors(book_pivot.iloc[237,:].values.reshape(1,-1),n_neighbors=6)
```

```python
distance
```

```python
suggestion
```

```python
for i in range(len(suggestion)):
  print(book_pivot.index[suggestion[i]])
```

```python
def recommend_book(book_name):
    book_id = np.where(book_pivot.index == book_name)[0][0]
    distance, suggestion = model.kneighbors(book_pivot.iloc[book_id,:].values.reshape(1,-1), n_neighbors = 6)
    for i in range(len(suggestion)):
        books = book_pivot.index[suggestion[i]]
        for j in books:
            print(j)
book_name = 'Harry Potter and the Chamber of Secrets (Book 2)'
recommend_book(book_name)
```
